Q18. What is an Optimizer? What are different types of optimizers? Explain each with an example.           
A18. An optimizer in machine learning is an algorithm used to adjust the weights of a model in order to minimize the loss function     
types of optimizers:
1. GRADIENT DESCENT (GD)       
- uses the full dataset to compute gradient and update parameters
- accurate updates   
- slow for large dataset     
- Example: pseudo-code
        for epoch in range(epochs):
            gradiants = compute_gradient(X_full, y_full)
            weights = weights - learning_rate * gradients   

2. STOCHASTIC GRADIENT DESCENT (SGD)
- Uses one random sample per update
- Faster than GD, but more noisy updates.
- Great for large datasets
- Examples:
            for epoch in range(epochs):
                for i in range(n_samples):
                    xi, yi = X[i], y[i]  
                    gradients = compute_gradients(xi, yi)
                    weights = weights - learning_rate * gradients         

3. MINI_BATCH GRADIENT DESCENT   
- Compromise between GD anf SGD
- Uses small batches for update    
- Widely used in practice     

4. MOMENTUM     
- Add a velocity term to remember the direction of the past gradients
- Helps accelerate and reduce oscillations
- Example:    
           optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)      

5. RMSprop    
- Divides learning rate by a moving average of a squared gradients
- Adapts learning rate for each parameter  
- Good for non-stationary objectives    
- Examples:
           optimizer = torch.optim.RMSprop(model.parameters(), lr=0.001)
